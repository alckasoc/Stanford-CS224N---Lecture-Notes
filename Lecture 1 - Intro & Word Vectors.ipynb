# 1. Introduction

What does "meaning" mean to us? 

- idea that is represented by a word/phrase
- idea that a person wants to express using words/signs
- idea that is expressed in a work of writing or art

We usually model the way we think of meaning as __denotational semantics__:

__signifier__ (symbol) $\Longleftrightarrow$ __signified__ (idea or thing)

__WordNet__ groups together words that are syonymous but is limited in capability.

In traditional NLP, we represent words as one-hot vectors. This is called a __localist representation__. However, using one-hot vectors to encode words lacks similarity between words and context in general. In the modern day, we use __embedding vectors__ built on the idea of __distributional semantics__ which defines a words meaning not by what idea it's associated with, but by what context or window of words the word appears in.

# 2. Word2Vec

__Word2vec__ is a 2013 framework for learning word vectors. 

The idea is to go through a large corpus/body of text and define a vocabulary table where each word is represented by an embedding vector. Go through each position $t$ in the text with center word $c$ and context words $o$ and calculate the probability of $o$ given $c$. We adjust the word vectors to maximize this probability.

 <br>
_Figure 2. Word2vec._

$$
Likelihood = L(\theta) = \prod_{t=1}^{T} \prod_{j \ne 0; -m \le j \le m} P(w_{t + j} ~|~ w_{t}; \theta) \hspace{1em} (Eq.~1)\\
Objective ~ Function = J(\theta) = \frac{1}{T}log L(\theta) = -\frac{1}{T} \sum_{t = 1}^{T} \sum_{j \ne 0; -m \le j \le m} log P(w_{t + j} ~|~ w_{t}; \theta)  \hspace{1em} (Eq.~2)\\
P(o ~|~ c) = \frac{exp(u_{o}^T v_c)}{\sum_{w \in V} exp(u_{w}^T v_c)} \hspace{1em} (Eq.~3)\\
$$

We keep 2 vectors per word: 1 vector for center word and 1 for context word. Notice Eq. 3 is like a softmax function. Also, remember that these vectors capture unique relationships between words. The vector for "king" is to "man" as "queen" is to "woman".
