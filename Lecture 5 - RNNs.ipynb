# 2. Language Models

__Language Modeling__ is the task of predicting what word comes next. 

$$
P(x^{(1)}, ..., x^{(T)}) = \prod_{t = 1}^{T} P(x^{(t)} ~|~ x^{(t - 1)}, ..., x^{(1)}) \hspace{1em} (Eq.~1)\\
$$

Pre-deep learning, we learned what was called an __n-gram Language model__. An n-gram is a chunk of $n$ consecutive words. We make a markov assumption that the word $x^{(t + 1)}$ depends on only the preceding $n - 1$ words.

These models suffered from a sparsity problem (i.e. what if it comes across an n-gram it has never seen before?). They also need to store the observed n-grams.
