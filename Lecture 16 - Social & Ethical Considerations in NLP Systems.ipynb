# 1. Introduction

Today's guest lecture focuses on:
* motivating examples as to why we should consider the ethics of NLP systems
* overview of topics in the intersection of ethics and NLP
* deep dive into 1 or 2 studies

__Ethics__ is the study of what are good and bad ends to pursue in life and what it is right and wrong to do in the conduct of life.

Ethics is not just black and white. There are many gray areas with obscure answers. Ethics is dynamic and fluctates with changes in values and beliefs  in society.

Consider an IQ classifier based on facial features. Building such a system without proper preparation may cause all sorts of problems. It may lead to prejudice. it may more than likely spit out incorrect results. It may be biased towards certain face archetypes.

 <br>
_Figure 1. "AI Gaydar" study._

The above study ran into many problems. The data they sourced were from a popular American dating website. These participants never agreed to the study. The dataset itself is highly biased as it is based off of who uses that app. There are also problems in choosing a deep learning model for this task. It may be costly to misclass certain instances and deep learning models are generally hard to interpret.

Lots of factors must be considered to evaluate the ethical side of NLP systems:
* algorithmic bias
* incivility
* privacy violation
* misinformation
* technological divide

We have a default social group in our head because of an implicit bias. We have 2 systems:
* System 1
    * is automatic and fast 
* System 2
    * slow, controlled, and rule-governed
    
We often make decisions based on system 1 when we associate ourselves with system 2. __Cognitive bias__ is a systematic pattern of deviation from rational judgment.

Here are just a few biases:
* confirmation bias
* ingroup favoritism
* group attribution error
* halo effect
* just-world hypothesis

How do these implicit biases in System 1 manifest?

They come in microaggressions, a comment/action that subtly and often unintentionally/unconsciously expresses a prejudiced attitude towards people of marginalized groups.

There are biases in language but also biases in datasets. Annotator's may be biased in annotating. Data selection and sampling may be biased.

Interestingly, SOTA NLP models can't identify microaggressions. They also don't incorporate social/cultural knowledge.

We need to transition from reactive approaches (seeing the bias in AI after they show it) to proactive approaches that will mitigate bias to begin with.
