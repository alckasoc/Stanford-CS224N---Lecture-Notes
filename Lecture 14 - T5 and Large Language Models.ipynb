# 1. Introduction

__Text-to-Text Transfer Transformer (T5)__ is a large language model (LLM). A lot of people are working in this field and it's difficult to see who's methods are doing better if so many variables are changing. This paper was on seeing how far they could take the current tools available. 

In the paper they built their own dataset (c4, available on TensorFlow) to test how pretraining helps.

_Figure 3. T5 Different pretraining objectives._

They found the last 2 options to be the most effective. Note, the last 2 options pretrain the model to predict spans of tokens while the first 2 options predict entire sequences.

 <br>
_Figure 4. T5 Different pretraining datasets._

They found that pretraining datasets more tailored towards a downstream task tended to yield a better model when it's finetuned.

 <br>
_Figure 5. T5 Different dataset sizes._

They found that your pretraining dataset needs to be large enough for the model to not overfit.

 <br>
_Figure 6. T5 multitask training._

They also tested multitask training and found that it would generally sacrifice some performance.

 <br>
_Figure 7. T5 multitask training and unsupervised pre-training + fine-tuning._

They found that if you do multi-task pre-training followed by fine-tuning, it doesn't sacrifice that much performance.

 <br>
_Figure 8. T5 model size and ensembling._

They found that longer training and larger models make for better performance!

Combining all these experiments together to form the best possible pipeline:
* encoder-decoder architecture
* span prediction objective
* C4 dataset
* multitask pre-training
* bigger models trained longer

 <br>
_Figure 9. Different model variants._

 <br>
_Figure 10. T5 Model variants on different downstream tasks._

Since T5 was trained on English, they also explored its applications in a multi-lingual setting.

They also introduced another task called the __closed-book question answering__. This is basically asking the model a question like "What color is a lemon?" without giving the model any context or database like in reading comprehension or open-domain QA tasks. This was for the authors to test how much knowledge the model actually learned.

They found larger models to "learn" more knowledge. But they still lagged behind in the SOTA for open-domain QA. They pretrained on __salient span masking__ which is basically masking mentions. They found by adding this objective to pretraining, they performed better than the SOTA in open-domain QA. The takeaway here is that pretraining for a downstream task should be done on a dataset relevant to that downstream task.

 <br>
_Figure 11. Memorization._

They found that larger models can memorize more. And models do, in fact, memorize!

They also explored different Transformer modifications.

 <br>
_Figure 12. Transformer modification validation losses._

They found that the original architecture is already pretty good. Only a few modifications (some expensive some minor) led to minor improvements.
