# 3. Compositional Representations and Systematic Generalization

The idea of __compositionality__ is that an expression is a function of the meaning of its parts. For example, "Brown cow" is the union of brown and cow. And Brown cow is exactly what combining brown and cow entails. However, English is not purely compositional. We have phrases like "red herring" which aren't literal red herrings, but more so symbolic phrases.

__Tree Reconstruction Error (TRE)__ is a loss for how well a model can reconstruct a dependency tree. Some work has shown that with learning, compositionality in the model improves as well.

 <br>
_Figure 4. Compositionality improvements._

The other question of interest here is: does the model generalize to realistic natural language?

Short answer: no!

# 4. Improving How We Evaluate NLP Models

 <br>
_Figure 5. Dynamic benchmarks._

Since language naturally is dynamic, we can have __dynamic benchmarks__ instead of static ones. This human-in-the-loop benchmark is time intensive.

 <br>
_Figure 6. Grounding benchmarks._

Some work has proposed that we need to ground language in different aspects not just web extracted data. There needs to be social aspects to it, meaning needs to be injected, etc.

# 5. Concluding Thoughts

In the short term, we need to see what works and doesn't work. Scaling clearly helps our models but what are other ways to improve? In the long term, we need to see how well they generalize and understand the ins and outs of language embedded in the context of our society. We need trustable and robust benchmarks to evaluate this.
