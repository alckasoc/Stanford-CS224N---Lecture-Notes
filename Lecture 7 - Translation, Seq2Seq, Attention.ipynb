# 1. Introduction

__Neural Machine Translation (NMT)__ is the task of having a machine translate text.

In the 20th century, people aimed to learn a transition model $P(x~|~y)$ from the data. We want to find the best English sentence $y$ given a French sentence $x$. 

 <br>
_Figure 1. Learning a Probabilistic model from data._

To learn the probabilistic model or translation model, we can break it down further and learn a model $P(x, a~|~y)$ where $a$ is the laatent variable. $a$ is the _alignment_ or the word-level correspondence between source sentence $x$ and target sentence $y$. There are some complications with this though. Alignments can be one-to-one, many-to-one, or many-to-many.

__Statistical Machine Translation__ was popular from the 1990s to the early 2010s and most work in this field are extremely complex, require lots of feature engineering, and lots of human effort to maintain.

__Neural Machine Translation__ does machine translation with 1 end-to-end neural network.

Sequence-to-sequence models (deep learning models) are useful for not just translation. They can be used for summarization, dialogue, parsing, etc.

__Greedy Decoding__ means the decoder will decode the same sentence every time. This might seem a little boring when a model is tasked with sentence generation. __Beam search__ keeps track of the $k$ most probable partial tanslations (called _hypotheses_). $k$ is the beam size which is usually set to somewhere between 5 and 10. 

NMT has many advantages compared to SMT:
- end-to-end
- fluent and has more context
- more powerful
- no feature engineering

however NMT also has some cons:
- hard to control and interpret

# 2. Evaluating Machine Translation

__Bilingual Evaluation Understudy (BLEU)__ compares the machine-written translation to one or several human-written translations. It computes a similarity score based on:
- n-gram precision
- penalty for too-short system translations

BLEU is imperfect, however. Your model can have high n-gram overlaps but poor performance. 

NMT is perhaps the most successful story of deep learning in NLP as it outperformed SMT methods that took years to build.

There are still many problems to NMT. 

- common sense is hard
- biases learned from data
- uninterpretable
